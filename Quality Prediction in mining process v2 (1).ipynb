```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
```

```python
file_path = r'C:\Users\harsh\Downloads\MiningProcess_Flotation_Plant_Database.csv'
```

```python
# Load the data, so we have to convert the date column into date 
# and we have to drop the duplicates entries/row using code below

data =  pd.read_csv(file_path,
                   decimal=",",
                    parse_dates=["date"],
                    infer_datetime_format=True).drop_duplicates()
data.info()
```

```python
#Check the shape of data (row and column)
data.shape
```

```python
#check the data if there any missing value or not
data.isnull().sum()
```

```python
#Display the data, and observe what kind of the data is this
data.head()
```

```python
# Obtain the correlation matrix
fig, ax = plt.subplots(figsize=(18, 18))
cor= data.corr()
sns.heatmap(cor, annot=True)
plt.title('Heatmap of Correlation between Features')
plt.show()
```

```python
#Drop data that there are no significant corelation on dependent feature
#Make Correlation with output variable
cor_target = abs(cor["% Silica Concentrate"])

#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.15]
relevant_features
```

```python
#We pick the 2 biggest corelation exclude target
relevant_features = relevant_features.nlargest(n=3)
```

```python
#Make a data from the relevant features
data = pd.DataFrame(data, columns=relevant_features.index)
data.head()
```

```python
#Checking The Outlier in our data
sns.boxplot(data['Flotation Column 01 Air Flow'])
```

```python
#Checking The Outlier in our data
sns.boxplot(data['% Iron Concentrate'])
```

```python
#Checking The Outlier in our data
sns.boxplot(data['% Silica Concentrate'])
```

```python
#Dropping the outlier with Percentiles
for i in data:
    upper_lim = data[i].quantile(.95)
    lower_lim = data[i].quantile(.05)

    data = data[(data[i] < upper_lim) & (data[i] > lower_lim)]
```

```python
# Before we split into train and test data, as we can see, the data have differents in units and magnitude
# So to make it at the same magnitude we can scaling the data

y = data['% Silica Concentrate']
X = data.drop(['% Silica Concentrate'], axis=1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
# After we scaled the data, and the data have the same magnitude
# we can split the data into Train & Test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled,
                                                    y,
                                                    test_size=0.3,
                                                   random_state=30)
```

###### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_pred_linreg = lin_reg.predict(X_test)
```

###### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
y_pred_ridge = ridge_regressor.predict(X_test)
```

###### Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X_train,y_train)
y_pred_lasso = lasso_regressor.predict(X_test)
```

###### XGBoost

```python
import xgboost as xgb
xgb = xgb.XGBRegressor(objective="reg:linear", random_state=42)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

###### Random forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split



# Create a RandomForestRegressor instance
random_forest = RandomForestRegressor()

# Define the hyperparameters grid
parameters = {
    'n_estimators': [10],  # Number of trees in the forest
    'max_depth': [10],   # Maximum depth of the tree
    
}

# Create the GridSearchCV object
random_forest_regressor = GridSearchCV(random_forest, parameters, scoring='neg_mean_squared_error', cv=5)


# Fit the model using GridSearchCV
random_forest_regressor.fit(X_train, y_train)

# Make predictions
y_pred_random_forest = random_forest_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

#LINEAR REGRESSION
MSE_linreg = mean_squared_error(y_test, y_pred_linreg)
print('Our Linear Regression mean squared error is: ',MSE_linreg)
MAE_linreg = mean_absolute_error(y_test, y_pred_linreg)
print('Our Linear Regression mean absolute error is: ',MAE_linreg)
R2_linreg = r2_score(y_test, y_pred_linreg) 
print('Our Linear Regression R2 score is: ', R2_linreg)
RMSE_linreg=np.sqrt(mean_squared_error(y_test, y_pred_linreg))
print('Our Linear Regreesion Root Mean Squared Error is:', np.sqrt(mean_squared_error(y_test, y_pred_linreg)))
print('-'*100)
print('-'*100)
#RIDGE REGRESSION
MSE_ridge = mean_squared_error(y_test, y_pred_ridge)
print('Our Rdige Regression mean squared error is: ',MSE_ridge)
MAE_ridge = mean_absolute_error(y_test, y_pred_ridge)
print('Our Ridge Regression mean absolute error is: ',MAE_ridge)
R2_ridge = r2_score(y_test, y_pred_ridge) 
print('Our Ridge Regression R2 score is: ', R2_ridge)
RMSE_ridge=np.sqrt(mean_squared_error(y_test, y_pred_ridge))
print('Our Ridge Regression Root Mean Squared Error is:', np.sqrt(mean_squared_error(y_test, y_pred_ridge)))
print('-'*100)
print('-'*100)
#LASSO REGRESSION
MSE_lasso = mean_squared_error(y_test, y_pred_lasso)
print('Our Lasso Regression mean squared error is: ',MSE_lasso)
MAE_lasso = mean_absolute_error(y_test, y_pred_lasso)
print('Our Lasso Regression mean absolute error is: ',MAE_lasso)
R2_lasso = r2_score(y_test, y_pred_lasso) 
print('Our Lasso Regression R2 score is: ', R2_lasso)
RMSE_lasso=np.sqrt(mean_squared_error(y_test, y_pred_lasso))
print('Our Lasso Regression Root Mean Squared Error is:', np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
print('-'*100)
print('-'*100)
#XGBOOST
MSE_xgb = mean_squared_error(y_test, y_pred_xgb)
print('Our XGBoost mean squared error is: ',MSE_xgb)
MAE_xgb = mean_absolute_error(y_test, y_pred_xgb)
print('Our XGBoost mean absolute error is: ',MAE_xgb)
R2_xgb = r2_score(y_test, y_pred_xgb) 
print('Our XGBoost R2 score is: ', R2_xgb)
RMSE_xgb= np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print('Our XGBoost Root Mean Squared Error is:', np.sqrt(mean_squared_error(y_test, y_pred_xgb)))
print('-'*100)
print('-'*100)
# RANDOM FOREST REGRESSION
MSE_random_forest = mean_squared_error(y_test, y_pred_random_forest)
print('Our Random Forest mean squared error is: ', MSE_random_forest)
MAE_random_forest = mean_absolute_error(y_test, y_pred_random_forest)
print('Our Random Forest mean absolute error is: ', MAE_random_forest)
R2_random_forest = r2_score(y_test, y_pred_random_forest) 
print('Our Random Forest R2 score is: ', R2_random_forest)
RMSE_random_forest =np.sqrt(mean_squared_error(y_test, y_pred_random_forest))
print('Our Random Forest Root Mean Squared Error is:', np.sqrt(mean_squared_error(y_test, y_pred_random_forest)))
```

### 4. Model Comparison

```python
import matplotlib.pyplot as plt

# Define model names
model_names = ['Linear Regression', 'Ridge Regression', 'Lasso Regression', 'XGBoost', 'Random Forest']

# Define corresponding evaluation metric scores (MSE, MAE, R2, RMSE)
MSE_scores = [MSE_linreg, MSE_ridge, MSE_lasso, MSE_xgb, MSE_random_forest]
MAE_scores = [MAE_linreg, MAE_ridge, MAE_lasso, MAE_xgb, MAE_random_forest]
R2_scores = [R2_linreg, R2_ridge, R2_lasso, R2_xgb, R2_random_forest]
RMSE_scores = [RMSE_linreg, RMSE_ridge, RMSE_lasso, RMSE_xgb, RMSE_random_forest]

# Create subplots for each evaluation metric in a 2x2 grid
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plotting MSE comparison
axes[0, 0].bar(model_names, MSE_scores, color='skyblue')
axes[0, 0].set_xlabel('Models')
axes[0, 0].set_ylabel('MSE')
axes[0, 0].set_title('Mean Squared Error Comparison')

# Plotting MAE comparison
axes[0, 1].bar(model_names, MAE_scores, color='lightgreen')
axes[0, 1].set_xlabel('Models')
axes[0, 1].set_ylabel('MAE')
axes[0, 1].set_title('Mean Absolute Error Comparison')

# Plotting R2 comparison
axes[1, 0].bar(model_names, R2_scores, color='salmon')
axes[1, 0].set_xlabel('Models')
axes[1, 0].set_ylabel('R2')
axes[1, 0].set_title('R-squared (R2) Comparison')

# Plotting RMSE comparison
axes[1, 1].bar(model_names, RMSE_scores, color='Brown')
axes[1, 1].set_xlabel('Models')
axes[1, 1].set_ylabel('RMSE')
axes[1, 1].set_title('Root Mean Square Error Comparison')

plt.tight_layout()
plt.show()

```

### 5. Tuning XG Boost Model

```python
# Hyper Parameter Tunning
from sklearn.model_selection import RandomizedSearchCV

params={
 "learning_rate"    : [0.01, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7, 0.9, 1.0 ]
    
}
```

```python
#Using Randomized Search CV to look the best parameter
random_search= RandomizedSearchCV(estimator=xgb,
                                param_distributions=params,
                                cv=5, n_iter=50,
                                scoring = 'r2',n_jobs = 4,
                                verbose = 1, 
                                return_train_score = True,
                                random_state=42)
```

```python
#Train Hyperparameter into our Data
random_search.fit(X_train, y_train)
```

```python
#Check the best estimator
random_search.best_estimator_
```

```python
#Using the best hyperparameter into our model
import xgboost as xgb
xgb = xgb.XGBRegressor(base_score=0.5, booster=None, colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1.0, gamma=0.2, gpu_id=-1,
             importance_type='gain', interaction_constraints=None,
             learning_rate=0.25, max_delta_step=0, max_depth=10,
             min_child_weight=7, monotone_constraints=None,
             n_estimators=100, n_jobs=0, num_parallel_tree=1,
             objective='reg:linear', random_state=42, reg_alpha=0, reg_lambda=1,
             scale_pos_weight=1, subsample=1, tree_method=None,
             validate_parameters=False, verbosity=None)
xgb.fit(X_train, y_train)
y_pred_xgb_tunning = xgb.predict(X_test)
```

```python
#Check Metrics after tunning
MSE = mean_squared_error(y_test, y_pred_xgb_tunning)
print('Our XGBoost after tunning mean squared error is: ',MSE)
MAE = mean_absolute_error(y_test, y_pred_xgb_tunning)
print('Our XGBoost after tunning mean absolute error is: ',MAE)
R2 = r2_score(y_test, y_pred_xgb_tunning) 
print('Our XGBoost after tunning R2 score is: ', R2)
print('Our XGBoost after tunning Root Mean Squared Error is:', np.sqrt(mean_squared_error(y_test, y_pred_xgb_tunning)))
```
